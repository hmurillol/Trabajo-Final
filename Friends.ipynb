<h3 style="color: #C71585">3.4 Binning</h3>

```python
dffriends1.head(1)
```

```python
dffriends1.dtypes
```

```python
dffriends1.dropna(subset = ["Año"], axis = 0, inplace = True)
```

```python
dffriends1[["Año"]] = dffriends1[["Año"]].astype("int", copy=True)
```

```python
dffriends1.dtypes
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

plt.pyplot.hist(dffriends1["Año"])

plt.pyplot.xlabel("Año")
plt.pyplot.ylabel("Cantidad")
plt.pyplot.title("Año bins")
```

```python
bins = np.linspace(min(dffriends1["Año"]), max(dffriends1["Año"]), 4)
bins
```

```python
group_names = ["bajo", "medio", "alto"]
dffriends1["Año-bins"] = pd.cut(dffriends1["Año"], bins, labels=group_names, include_lowest=True)
dffriends1[["Año", "Año-bins"]].head(3)
```

```python
dffriends1["Año-bins"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

plt.pyplot.hist(dffriends1["Año-bins"])

plt.pyplot.xlabel("Año")
plt.pyplot.ylabel("cantidad")
plt.pyplot.title("Año bins")
```
